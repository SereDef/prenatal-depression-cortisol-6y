# Prental depression exposure and hair cortisol at age 6

```python
i = pd.concat([int_only.mean(0), int_only.var(0)/2], axis =1); i.columns = ["im", "ist"]
m = pd.concat([multimorbid.mean(0), multimorbid.var(0)/2], axis =1); m.columns = ["mm", "mst"]
h = pd.concat([healthy.mean(0), healthy.var(0)/2], axis =1); h.columns = ["hm", "hst"]
f = pd.concat([fm_only.mean(0), fm_only.var(0)/2], axis =1); f.columns = ["fm", "fst"]

sim = pd.concat([i, m, h, f], axis =1)
sim['diff'] = sim.mm - sim.im
sim = sim.sort_values(by = 'diff') 
```
